```python
# integer variable
# Python supports arbitrarily large integers, so you can use very large numbers without worrying about overflow
# The following is an example of a very large integer
a = 9643785489534875845384656734856783465348768
print(a)

# float variable
# Python supports floating-point numbers, which can represent real numbers with decimal points
b = 3.14159
print(b)

# complex variable
# Python supports complex numbers, which have a real part and an imaginary part
c = 2 + 3j
print(c)

# string variable
# Strings in Python are sequences of characters enclosed in single or double quotes
d = "Hello, World!"
print(d)

# boolean variable
# Boolean variables can be either True or False
e = True
print(e)

# list variable
# Lists in Python are ordered collections of items that can be of different types
f = [1, 2, 3, "four", 5.0, True]
print(f)

# tuple variable
# Tuples are similar to lists but are immutable (cannot be changed after creation)
g = (1, 2, 3, "four", 5.0, True)
print(g)

# dictionary variable
# Dictionaries are collections of key-value pairs, where each key is unique
h = {"name": "Kunal", "age": 21, "university": "IIT Jammu"}
print(h)

# set variable
# Sets are unordered collections of unique items
i = {1, 2, 3, 4, 5}
print(i)

# printing types of variables
print("a has a type of ", type(a))
print("b has a type of ", type(b))
print("c has a type of ", type(c))
print("d has a type of ", type(d))
print("e has a type of ", type(e))
print("f has a type of ", type(f))
print("g has a type of ", type(g))
print("h has a type of ", type(h))
print("i has a type of ", type(i))
```

```python
# Operators in Python
# Arithmetic Operators
x = 10
y = 5
print("Addition:", x + y)          # Addition
print("Subtraction:", x - y)       # Subtraction
print("Multiplication:", x * y)    # Multiplication
print("Division:", x / y)          # Division
print("Floor Division:", x // y)    # Floor Division
print("Modulus:", x % y)            # Modulus
print("Exponentiation:", x ** y)    # Exponentiation

# Comparison Operators
print("Equal:", x == y)              # Equal
print("Not Equal:", x != y)          # Not Equal
print("Greater than:", x > y)        # Greater than
print("Less than:", x < y)           # Less than
print("Greater than or equal to:", x >= y)  # Greater than or equal to
print("Less than or equal to:", x <= y)      # Less than or equal to

# Logical Operators
print("Logical AND:", x > 5 and y < 10)  # Logical AND
print("Logical OR:", x > 5 or y < 3)    # Logical OR
print("Logical NOT:", not (x > 5))       # Logical NOT

# Bitwise Operators
print("Bitwise AND:", x & y)            # Bitwise AND
print("Bitwise OR:", x | y)             # Bitwise OR
print("Bitwise XOR:", x ^ y)            # Bitwise XOR
print("Bitwise NOT:", ~x)                # Bitwise NOT
print("Left Shift:", x << 1)            # Left Shift
print("Right Shift:", x >> 1)           # Right Shift

# Assignment Operators
x = 10
x += 5  # x = x + 5
print("After += :", x)
x -= 3  # x = x - 3
print("After -= :", x)
x *= 2  # x = x * 2
print("After *= :", x)
x /= 4  # x = x / 4
print("After /= :", x)
x //= 2  # x = x // 2
print("After //= :", x)
x %= 3  # x = x % 3
print("After %= :", x)
x **= 2  # x = x ** 2
print("After **= :", x)

# Identity Operators
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print("a is b:", a is b)          # True, because b is a reference to a
print("a is c:", a is c)          # False, because c is a different object with the same content
print("a is not c:", a is not c)
print("b is c:", b is c)          # False, because b and c are different objects
print("a is not b:", a is not b)  # False, because b is a reference to a

# Membership Operators
my_list = [1, 2, 3, 4, 5]
print("1 in my_list:", 1 in my_list)      # True, because 1 is an element of my_list
print("6 in my_list:", 6 in my_list)      # False, because 6 is not an element of my_list
print("1 not in my_list:", 1 not in my_list)  # False, because 1 is in my_list
print("6 not in my_list:", 6 not in my_list)    # True, because 6 is not in my_list
```

```python
# Conditionals in Python
# Conditional statements allow you to execute code based on certain conditions
x = input("Enter a number: ")
x = int(x)  # Convert input to integer for comparison
if x > 0:
    print("x is positive")
elif x < 0:
    print("x is negative")
else:
    print("x is zero")

# Nested conditionals
y = input("Enter another number: ")
y = int(y)  # Convert input to integer for comparison
if x > 0:
    if y > 0:
        print("Both x and y are positive")
    elif y < 0:
        print("x is positive, but y is negative")
    else:
        print("x is positive, but y is zero")
elif x < 0:
    if y > 0:
        print("x is negative, but y is positive")
    elif y < 0:
        print("Both x and y are negative")
    else:
        print("x is negative, but y is zero")
else:
    if y > 0:
        print("x is zero, but y is positive")
    elif y < 0:
        print("x is zero, but y is negative")
    else:
        print("Both x and y are zero")
```

```python
def match_case_example(value):
    match value:
        case 0:
            return "Zero"
        case int() as number if number > 0:
            return f"Positive integer: {number}"
        case int() as number if number < 0:
            return f"Negative integer: {number}"
        case float():
            return f"Float: {value}"
        case str() as s if s.isdigit():
            return f"String containing digits: {s}"
        case [first, second, *rest]:
            return f"List with at least two elements: {first}, {second}, rest: {rest}"
        case {'name': n, 'age': a}:
            return f"Dictionary with name and age: {n}, {a}"
        case _:
            return "No match found"

# Example usages
print(match_case_example(0))                    # Zero
print(match_case_example(42))                   # Positive integer: 42
print(match_case_example(-7))                   # Negative integer: -7
print(match_case_example(3.14))                 # Float: 3.14
print(match_case_example("12345"))              # String containing digits: 12345
print(match_case_example([1, 2, 3, 4]))         # List with at least two elements: 1, 2, rest: [3, 4]
print(match_case_example({'name': 'Kunal', 'age': 21}))  # Dictionary with name and age: Kunal, 21
print(match_case_example("hello"))              # No match found
```

```python
# For loops in Python
# For loops are used to iterate over a sequence (like a list, tuple, or string)
# Example: Iterating over a list
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print("Number:", num)

# Example: Iterating over a string
for char in "Hello":
    print("Character:", char)

# Example: Using range() function in a for loop
for i in range(5):
    print("Range value:", i)

# Example: Nested for loops
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")

# Example: Using else with a for loop
for i in range(0,3):
    print("Current value:", i)
else:
    print("Loop completed without break")
```

```python
# Simple while loop: prints count from 0 to 4
count = 0
while count < 5:
    print("Count:", count)
    count += 1

# While loop with break and checking even/odd
count = 0
while count < 10:
    if count == 5:
        print("Breaking the loop at count =", count)
        break  # Exit the loop when count is 5
    if count % 2 == 0:
        print("Even count:", count)
    else:
        print("Odd count:", count)
    count += 1

# While loop using continue to skip even numbers
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue  # Skip the rest of the loop for even numbers
    print("Odd count:", count)

# While loop with else: else block runs when loop condition becomes false (not by break)
count = 0
while count < 5:
    print("Count:", count)
    count += 1
else:
    print("Reached the end of the while loop, count is now:", count)
```

```python
# Functions in Python
# Functions are reusable blocks of code that perform a specific task
# Defining a simple function
def fib(n):
    """Function to return the nth Fibonacci number."""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

# Calling the function
print("Fibonacci of 5:", fib(5))  # Output: 5

# Function with default parameters
def greet(name="Guest"):
    """Function to greet a person."""
    return f"Hello, {name}!"

# Calling the function with and without parameters
print(greet("Kunal"))  # Output: Hello, Kunal!
print(greet())         # Output: Hello, Guest!

# Function with variable number of arguments
def add_numbers(*args):
    """Function to add any number of numbers."""
    return sum(args)

# Calling the function with different numbers of arguments
print("Sum of 1, 2, 3:", add_numbers(1, 2, 3))  # Output: 6
print("Sum of 4, 5, 6, 7:", add_numbers(4, 5, 6, 7))  # Output: 22

# Function with keyword arguments
def print_info(name, age, **kwargs):
    """Function to print name, age, and additional info."""
    info = f"Name: {name}, Age: {age}"
    for key, value in kwargs.items():
        info += f", {key}: {value}"
    return info

# Calling the function with keyword arguments
print(print_info("Kunal", 21, university="IIT Jammu", major="Computer Science"))    # Output: Name: Kunal, Age: 21, university: IIT Jammu, major: Computer Science
```
