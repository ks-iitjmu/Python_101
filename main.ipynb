```python
# Match Case Example
def match_case_example(value):
    match value:
        case 0:
            return "Zero"
        case int() as number if number > 0:
            return f"Positive integer: {number}"
        case int() as number if number < 0:
            return f"Negative integer: {number}"
        case float():
            return f"Float: {value}"
        case str() as s if s.isdigit():
            return f"String containing digits: {s}"
        case [first, second, *rest]:
            return f"List with at least two elements: {first}, {second}, rest: {rest}"
        case {'name': n, 'age': a}:
            return f"Dictionary with name and age: {n}, {a}"
        case _:
            return "No match found"

# Example usages
print(match_case_example(0))                    # Zero
print(match_case_example(42))                   # Positive integer: 42
print(match_case_example(-7))                   # Negative integer: -7
print(match_case_example(3.14))                 # Float: 3.14
print(match_case_example("12345"))              # String containing digits: 12345
print(match_case_example([1, 2, 3, 4]))         # List with at least two elements: 1, 2, rest: [3, 4]
print(match_case_example({'name': 'Kunal', 'age': 21}))  # Dictionary with name and age: Kunal, 21
print(match_case_example("hello"))              # No match found
```

```python
# Simple while loop: prints count from 0 to 4
count = 0
while count < 5:
    print("Count:", count)
    count += 1

# While loop with break and checking even/odd
count = 0
while count < 10:
    if count == 5:
        print("Breaking the loop at count =", count)
        break  # Exit the loop when count is 5
    if count % 2 == 0:
        print("Even count:", count)
    else:
        print("Odd count:", count)
    count += 1

# While loop using continue to skip even numbers
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue  # Skip the rest of the loop for even numbers
    print("Odd count:", count)

# While loop with else: else block runs when loop condition becomes false (not by break)
count = 0
while count < 5:
    print("Count:", count)
    count += 1
else:
    print("Reached the end of the while loop, count is now:", count)
```

```python
# Functions in Python
# Functions are reusable blocks of code that perform a specific task
# Defining a simple function
def fib(n):
    """Function to return the nth Fibonacci number."""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

# Calling the function
print("Fibonacci of 5:", fib(5))  # Output: 5

# Function with default parameters
def greet(name="Guest"):
    """Function to greet a person."""
    return f"Hello, {name}!"

# Calling the function with and without parameters
print(greet("Kunal"))  # Output: Hello, Kunal!
print(greet())         # Output: Hello, Guest!

# Function with variable number of arguments
def add_numbers(*args):
    """Function to add any number of numbers."""
    return sum(args)

# Calling the function with different numbers of arguments
print("Sum of 1, 2, 3:", add_numbers(1, 2, 3))  # Output: 6
print("Sum of 4, 5, 6, 7:", add_numbers(4, 5, 6, 7))  # Output: 22

# Function with keyword arguments
def print_info(name, age, **kwargs):
    """Function to print name, age, and additional info."""
    info = f"Name: {name}, Age: {age}"
    for key, value in kwargs.items():
        info += f", {key}: {value}"
    return info

# Calling the function with keyword arguments
print(print_info("Kunal", 21, university="IIT Jammu", major="Computer Science"))    # Output: Name: Kunal, Age: 21, university: IIT Jammu, major: Computer Science
```

```python
# Recursion in Python
# Recursion is a programming technique where a function calls itself to solve a problem
def factorial(n):
    """Function to calculate the factorial of a number using recursion."""
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Calling the factorial function
print("Factorial of 5:", factorial(5))  # Output: 120
print("Factorial of 0:", factorial(0))  # Output: 1

def fibonacci(n):
    """Function to return the nth Fibonacci number using recursion."""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Calling the fibonacci function
print("Fibonacci of 5:", fibonacci(5))  # Output: 5
```

```python
# Sets in Python
# Sets are unordered collections of unique items
# Creating a set
my_set = {1, 2, 3, "four", 5.0, True}
print("My Set:", my_set)
# Accessing elements in a set
# Sets do not support indexing, so we cannot access elements by index
# However, we can iterate over the set
for item in my_set:
    print("Set item:", item)

# Adding elements to a set
my_set.add("six")  # Adding an element to the set
print("Set after adding 'six':", my_set)

# Removing elements from a set
my_set.remove(2)  # Removing an element from the set
print("Set after removing 2:", my_set)
my_set.discard("four")  # Discarding an element (does not raise an error if not found)
print("Set after discarding 'four':", my_set)

# Popping an element from a set
popped_element = my_set.pop()  # Popping an arbitrary element from the set
print("Popped element:", popped_element)
print("Set after popping an element:", my_set)

# Clearing a set
my_set.clear()  # Removing all elements from the set
print("Set after clearing:", my_set)

# Set Operations
# Sets support various operations like union, intersection, difference, and symmetric difference
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print("Set A:", set_a)
print("Set B:", set_b)
print("Union of A and B:", set_a.union(set_b))  # Union of two sets
print("Intersection of A and B:", set_a.intersection(set_b))  # Intersection of two sets
print("Difference of A and B (A - B):", set_a.difference(set_b))  # Difference of two sets
print("Symmetric difference of A and B:", set_a.symmetric_difference(set_b))  # Symmetric difference of two sets

# Set Comprehensions
# Set comprehensions provide a concise way to create sets
squared_set = {x**2 for x in range(10)}  # Creating a set of squared numbers from 0 to 9
print("Squared Set:", squared_set)
# Filtering a set using set comprehension
even_set = {x for x in range(20) if x % 2 == 0}  # Creating a set of even numbers from 0 to 19
print("Even Set:", even_set)
```

```python
# Dictionaries in Python
# Dictionaries are collections of key-value pairs, where each key is unique
# Creating a dictionary
my_dict = {"name": "Kunal", "age": 21, "university": "IIT Jammu"}
print("My Dictionary:", my_dict)
# Accessing values in a dictionary
print("Name:", my_dict["name"])  # Accessing the value associated with the key
print("Age:", my_dict["age"])  # Accessing the value associated with the key
print("University:", my_dict["university"])  # Accessing the value associated with the key
# Adding key-value pairs to a dictionary
my_dict["major"] = "Computer Science"  # Adding a new key-value pair
print("Dictionary after adding major:", my_dict)
# Removing key-value pairs from a dictionary
my_dict.pop("age")  # Removing a key-value pair by key
print("Dictionary after removing age:", my_dict)
# Checking if a key exists in a dictionary
print("Is 'name' in my_dict?", "name" in my_dict)  # True
print("Is 'age' in my_dict?", "age" in my_dict)  # False
# Iterating over a dictionary
for key, value in my_dict.items():
    print(f"Key: {key}, Value: {value}")  # Iterating over key-value pairs

# Dictionary Methods
# Dictionaries have various built-in methods for manipulation
print("Keys in my_dict:", my_dict.keys())  # Getting all keys in the dictionary
print("Values in my_dict:", my_dict.values())  # Getting all values in the dictionary
print("Items in my_dict:", my_dict.items())  # Getting all key-value pairs in the dictionary
# Copying a dictionary
copied_dict = my_dict.copy()  # Creating a shallow copy of the dictionary
print("Copied Dictionary:", copied_dict)
# Clearing a dictionary
my_dict.clear()  # Removing all key-value pairs from the dictionary
print("Dictionary after clearing:", my_dict)
# Merging dictionaries
dict_a = {"a": 1, "b": 2}
dict_b = {"b": 3, "c": 4}
merged_dict = {**dict_a, **dict_b}  # Merging two dictionaries
print("Merged Dictionary:", merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}

# Dictionary Comprehensions
# Dictionary comprehensions provide a concise way to create dictionaries
squared_dict = {x: x**2 for x in range(5)}  # Creating a dictionary of squared numbers from 0 to 4
print("Squared Dictionary:", squared_dict)
# Filtering a dictionary using dictionary comprehension
even_dict = {x: x**2 for x in range(10) if x % 2 == 0}  # Creating a dictionary of even numbers and their squares from 0 to 9
print("Even Dictionary:", even_dict)
# Dictionary with default values
from collections import defaultdict
default_dict = defaultdict(int)  # Creating a defaultdict with default value of int (0)
default_dict["a"] = 1
default_dict["b"] = 2
print("Default Dictionary:", default_dict)  # Accessing a non-existing key returns the default value

# Nested Dictionaries
# Dictionaries can contain other dictionaries, creating a nested structure
nested_dict = {    "student1": {"name": "Kunal", "age": 21, "university": "IIT Jammu"},
    "student2": {"name": "Amit", "age": 22, "university": "IIT Delhi"},
    "student3": {"name": "Ravi", "age": 20, "university": "IIT Bombay"}
}
print("Nested Dictionary:", nested_dict)
# Accessing elements in a nested dictionary
print("Name of student1:", nested_dict["student1"]["name"])  # Accessing the name of student1
print("Age of student2:", nested_dict["student2"]["age"])  # Accessing the age of student2
print("University of student3:", nested_dict["student3"]["university"])  # Accessing the university of student3
# Iterating over a nested dictionary
for student, info in nested_dict.items():
    print(f"{student} - Name: {info['name']}, Age: {info['age']}, University: {info['university']}")  # Iterating over students and their information

# Dictionary with mixed types
# Dictionaries can have keys and values of different types
mixed_dict = {1: "one", "two": 2, 3.0: "three", (4, 5): "tuple_key"}
print("Mixed Dictionary:", mixed_dict)
# Accessing elements in a mixed dictionary
print("Value for key 1:", mixed_dict[1])  # Accessing the value for key 1
print("Value for key 'two':", mixed_dict["two"])  # Accessing the value for key 'two'
print("Value for key 3.0:", mixed_dict[3.0])  # Accessing the value for key 3.0
print("Value for key (4, 5):", mixed_dict[(4, 5)])  # Accessing the value for key (4, 5)

# Dictionary with complex keys
# Dictionaries can have complex keys like tuples or frozensets
# frozenset is an immutable version of a set, which can be used as a key
complex_dict = {("a", "b"): 1, frozenset([1, 2, 3]): 2}
print("Complex Dictionary:", complex_dict)
# Accessing elements in a complex dictionary
print("Value for key ('a', 'b'):", complex_dict[("a", "b")])  # Accessing the value for key ('a', 'b')
print("Value for key frozenset([1, 2, 3]):", complex_dict[frozenset([1, 2, 3])])  # Accessing the value for key frozenset([1, 2, 3])
```

```python
# Exceptions in Python
# Exceptions are errors that occur during the execution of a program
# Handling exceptions allows you to gracefully handle errors and continue execution
try:
    # Code that may raise an exception
    result = 10 / 0  # This will raise a ZeroDivisionError
except ZeroDivisionError as e:
    # Handling the specific exception
    print("Error: Division by zero is not allowed.", e)
except Exception as e:
    # Handling any other exception
    print("An unexpected error occurred:", e)

# Finally block to execute code regardless of whether an exception occurred
try:
    file = open("non_existent_file.txt", "r")  # Trying to open a non-existent file
except FileNotFoundError as e:
    print("Error: File not found.", e)
finally:
    print("This block always executes, regardless of whether an exception occurred or not.")

# Raising exceptions manually
def check_positive(number):
    """Function to check if a number is positive."""
    if number < 0:
        raise ValueError("Number must be positive")  # Raising a ValueError if the number is negative
    return number

try:
    print("Checking number:", check_positive(-5))  # This will raise an exception
except ValueError as e:
    print("Caught an exception:", e)

# Custom Exception
class CustomError(Exception):
    """Custom exception class."""
    pass

def raise_custom_error():
    """Function to raise a custom exception."""
    raise CustomError("This is a custom error message")

try:
    raise_custom_error()  # This will raise a CustomError
except CustomError as e:
    print("Caught a custom exception:", e)

# Using try-except-else
# The else block runs if no exceptions were raised in the try block
try:    
    result = 10 / 2  # This will not raise an exception
except ZeroDivisionError as e:
    print("Error: Division by zero is not allowed.", e)
else:
    print("Result:", result)

# Using try-except-finally
# The finally block runs regardless of whether an exception was raised or not
try:
    file = open("example.txt", "w")  # Trying to open a file for writing
    file.write("Hello, World!")  # Writing to the file
except IOError as e:
    print("Error: Unable to write to file.", e)
finally:
    print("This block always executes, regardless of whether an exception occurred or not.")
    if 'file' in locals():
        file.close()
```

```python
# Short hand If - Else - Statement
# Python allows a shorthand way to write if-else statements
a = 10
b = 20
max_value = a if a > b else b  # Assigning the maximum value to max_value
print("Maximum value is:", max_value)  # Output: Maximum value is: 20
```

```python
# Enumerate Functions in Python
# The enumerate() function adds a counter to an iterable and returns it as an enumerate object
# This is useful when you need both the index and the value of items in a loop
my_list = ['apple', 'banana', 'cherry']
for index, value in enumerate(my_list):
    print(f"Index: {index}, Value: {value}")
```

```python
# OS module in Python
# The os module provides a way to interact with the operating system
import os
# Getting the current working directory
current_directory = os.getcwd()
print("Current Working Directory:", current_directory)
# Listing files and directories in the current directory
files_and_dirs = os.listdir(current_directory)
print("Files and Directories in Current Directory:", files_and_dirs)
# Creating a new directory
new_directory = os.path.join(current_directory, "new_folder")
os.makedirs(new_directory, exist_ok=True)  # Creating a new directory if it doesn't exist
print("New Directory Created:", new_directory)
# Changing the current working directory
os.chdir(new_directory)  # Changing the current working directory to the new directory
print("Changed Current Working Directory to:", os.getcwd())
# Deleting the created directory
os.chdir(current_directory)  # Changing back to the original directory
os.rmdir(new_directory)  # Removing the new directory
print("Removed New Directory:", new_directory)
# Importing the sys module
# The sys module provides access to some variables used or maintained by the interpreter
import sys
# Getting the Python version
python_version = sys.version
print("Python Version:", python_version)
# Getting the platform information
platform_info = sys.platform
print("Platform Information:", platform_info)
# Getting the command line arguments
command_line_args = sys.argv
print("Command Line Arguments:", command_line_args)  # This will print the script name and any arguments passed to the script
# # Exiting the program
# sys.exit("Exiting the program")  # Exiting the program with a message
# # Note: The sys.exit() will terminate the script, so any code after this line will not be executed.
```
